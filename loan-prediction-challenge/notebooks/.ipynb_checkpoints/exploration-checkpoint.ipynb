```python
# Unique values
def unique_values(dataframe, data_type='object'):
    for i in dataframe.select_dtypes(include=data_type).columns:
        print('Column: {}, Unique: {}'.format(
            i, len(dataframe.select_dtypes(include=data_type)[i].unique())))

# Convert to datetime
def convert_to_datetime(dataframe):
    '''Converts to datetime format.'''
    for column, value in zip(dataframe.columns,
                             dataframe.columns.str.contains('date')):
        if value == True:
            dataframe[column] = pd.to_datetime(dataframe[column],
                                               yearfirst=True,
                                               exact=False)
            
# Convert to category data type
def convert_to_category_type(dataframe):
    '''Converts object data to category data type.'''
    for column in dataframe.select_dtypes(include='object').columns:
        dataframe[column] = dataframe[column].astype('category')

# Label encoding
def label_encoding(dataframe):
    '''Encoding categorical variables.'''
    label_encoder = preprocessing.LabelEncoder()
    for column in dataframe.select_dtypes(include='category').columns:
        dataframe[column] = label_encoder.fit_transform(dataframe[column])

# Text formatting
def title_case(dataframe):
    _cols = (i for i in dataframe.select_dtypes(include='object').columns)
    try:
        while True:
            _col = _cols.__next__()
            dataframe[_col] = list(map(lambda i: i.title(), dataframe[_col]))
    except StopIteration:
        print('Done.\n')
    except KeyError:
        print('Nothing to do.\n')

# Frequency plot
def frequency_plot(dataframe, x, title, hue=None):
    '''Duel axis frequency plot with percentages.'''
    # Plot settings
    plt.rcParams['figure.figsize'] = (16.0, 9.0)
    # Size of dataframe
    ncount = len(dataframe)
    # Default chart
    ax = sns.countplot(data=dataframe, x=x, hue=hue)
    plt.title(title, fontsize=20)
    plt.xlabel(' ')
    # Make twin axes
    ax2=ax.twinx()
    # Switch axes - percentages on the left, frequency on the right
    ax2.yaxis.tick_right()
    ax.yaxis.tick_left()
    # Switch labels as well
    ax2.set_ylabel('Percentage', fontsize=16)
    ax.set_ylabel('Frequency', fontsize=16)
    for p in ax.patches:
        x = p.get_bbox().get_points()[:, 0]
        y = p.get_bbox().get_points()[1, 1]
        # Setting text alignment
        ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y),
                    ha='center', va='bottom')
    # Use aa LinearLocator to ensure the correct number of ticks
    ax.yaxis.set_major_locator(ticker.LinearLocator(11))
    # y-axis ticks
    ax2.yaxis.set_ticks(np.arange(0, 101, 10))
    ax2.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f%%'))
    # Axes range
    ax2.set_ylim(0, 100)
    ax.set_ylim(0, ncount)
    # Use a MultipleLocator to ensure a ticker spacing of 10
    ax2.yaxis.set_major_locator(ticker.MultipleLocator(10))
    # Grid lines
    ax2.grid(None)
    
# Calculate correlation coefficients
def corrfunc(x, y, **kws):
    '''Calculate correlation coefficients.'''
    r, _ = stats.pearsonr(x, y)
    ax = plt.gca()
    ax.annotate("r = {:.2f}".format(r),
                xy=(.1, .6), xycoords=ax.transAxes,
                size=24)
```

<a id='transforming-the-data'></a>
### Transforming

<a id='combine-the-datasets'></a>
#### Combine the datasets

```python
# Combine the datasets
traindemoperf = pd.merge(traindemographics, trainperf, on='customerid', how='outer')
train = pd.merge(traindemoperf, trainprevloans, on='customerid', how='outer')
```

<a id='convert-to-datetime'></a>
#### Convert to `datetime` format

```python
# Convert to datetime
convert_to_datetime(train)
```

<a id='missing-values-object-type'></a>
#### Missing values for `object` variables

```python
# Fill missing values for object data
for column in train.select_dtypes(include='object').columns:
    train[column].fillna(value='Other', inplace=True)

# Assert
assert train.select_dtypes(include='object').isnull().sum().values.sum() == 0
```

<a id='convert-to-category'></a>
#### Convert object variables to categorical data type

```python
# Convert to category data type
convert_to_category_type(train)
```

<a id='missing-values-int-float'></a>
#### Missing values for `int` and `float` data type

```python
# Fill missing values with the median
for column in train.select_dtypes(include=['int', 'float']).columns:
    train[column].fillna(value=train[column].median(), inplace=True)

# Assert
assert train.select_dtypes(include=['int', 'float']).isnull().sum().values.sum() == 0
```

<a id='missing-values-datetime'></a>
#### Missing values `datetime` data type

```python
# Use most common value
for column in train.select_dtypes(include='datetime').columns:
    train[column].fillna(value=train[column].mode()[0], inplace=True)

# Assert
assert train.isnull().sum().values.sum() == 0
```

```python
title_case(train)
```

### Visualizations

```python
# Frequency plot
for column in ['bank_account_type', 'employment_status_clients', 'level_of_education_clients', 'good_bad_flag']:
    frequency_plot(dataframe=train, x=column, title=column.title())
    plt.show()
```

We added the `other` placeholder to our dataset so that we can work with the data without worrying to much about missing values.

Let's see what we can learn from the placeholder value in the categorical data starting with our target variable.

```python
target_placeholder = train.loc[train['good_bad_flag'] == 'Other']
target_placeholder.shape
```

There over a thousand individuals whose loan does not have a rating.

**Who are these people?**

```python
fig, axes = plt.subplots(2, 1, sharex=True)
target_placeholder['employment_status_clients'].value_counts().plot.barh(figsize=(8, 5), ax=axes[0])
target_placeholder['level_of_education_clients'].value_counts().plot.barh(figsize=(8, 5), ax=axes[1])
axes[0].set_title('Employment Status of Unflagged Loan Borrowers', fontsize=16)
axes[1].set_title('Education Level of Unflagged Loan Borrowers', fontsize=16)
plt.tight_layout()
```

The majority are permanent employees or do not have an employment status. Curiously, the majority do not have a recorded education level.

Let's go back and look at the full dataset. We want to get a good overview of it.

**How are the loan amounts distributed?**

```python
# Data
data = train.loc[~train['good_bad_flag'].str.contains('Other')]

# Plot
fig, axes = plt.subplots(2, 2, figsize=(24, 20))
sns.categorical.boxplot(x='level_of_education_clients',
                        y='loanamount_x', 
                        hue='good_bad_flag', 
                        data=data,
                        ax=axes[0][0])
sns.categorical.boxplot(x='level_of_education_clients', 
                        y='loanamount_y', 
                        hue='good_bad_flag', 
                        data=data,
                        ax=axes[1][0])
sns.kdeplot(data['loanamount_x'], ax=axes[0][1])
sns.kdeplot(data['loanamount_y'], ax=axes[1][1])
axes[0][0].set_xlabel(''), axes[0][1].set_xlabel('')
axes[1][0].set_xlabel(''), axes[1][1].set_xlabel('')
axes[0][0].set_ylabel(r'Loan Amount ($x$)', fontsize=20)
axes[0][1].set_ylabel('')
axes[1][0].set_ylabel(r'Loan Amount ($y$)', fontsize=20)
axes[1][1].set_ylabel('')
axes[0][0].set_title('How are loans distributed by education level?', fontsize=30)
axes[0][1].set_title('Loan Amount Distribution', fontsize=30)
plt.tight_layout()
```

The loan amounts appear to have multiple distributions.

The level of one's education plays a role in whether they receive a loan or not. Candidates with lower education levels received lower value loans, while graduates and post-graduate loan candidates receive higher loan amounts.

Also, a higher level of education had greater variation in loan amounts, while lower levels of education had more compact distributions.

Candidates who indicated their level of education had more "good" flags than those who did not specify their education level.

**How many loans do customers have?**

```python
fig, axes = plt.subplots(1, 2)
sns.kdeplot(train.loannumber_x, ax=axes[0])
sns.kdeplot(train.loannumber_y, ax=axes[0])
sns.kdeplot(train.loannumber_x, train.loannumber_y, ax=axes[1])
axes[0].legend([r'Loan Number, $x$', r'Loan Number, $y$'])
axes[0].set_xlabel(r'Loan Number, $(x, y)$', fontsize=16)
axes[0].set_ylabel('Density', fontsize=16)
axes[1].set_xlabel(r'Loan Number, $x$', fontsize=16)
axes[1].set_ylabel(r'Loan Number, $y$', fontsize=16)
axes[0].set_title('How many loans do customers have?', fontsize=20)
axes[1].set_title('Loan Number Distribution', fontsize=20)
```

Most customer have less than 5 loans that they are servicing. Previous loan numbers had a lower density than latter loan numbers, with latter loans peaking in density at 8 loans per customer.

```python
fig, axes = plt.subplots(2, 2, figsize=(24, 20), sharex=True)
sns.barplot(x='bank_name_clients', y='loanamount_x', data=train, ax=axes[0][0])
sns.barplot(x='bank_name_clients', y='loanamount_y', data=train, ax=axes[1][0])
sns.violinplot(x='bank_name_clients', y='totaldue_x', data=train, ax=axes[0][1])
sns.violinplot(x='bank_name_clients', y='totaldue_y', data=train, ax=axes[1][1])
axes[0][0].set_xlabel(''), axes[0][1].set_xlabel('')
axes[1][0].set_xlabel(''), axes[1][1].set_xlabel('')
axes[0][0].set_ylabel(r'Loan Amount ($x$)', fontsize=20)
axes[1][0].set_ylabel(r'Loan Amount ($y$)', fontsize=20)
axes[0][1].set_ylabel(r'Amount Due ($x$)', fontsize=20)
axes[1][1].set_ylabel(r'Amount Due ($y$)', fontsize=20)
axes[1][0].tick_params(axis='x', labelrotation=90, labelsize=20)
axes[1][1].tick_params(axis='x', labelrotation=90, labelsize=20)
plt.tight_layout()
```

Keystone Bank offered the greatest number of latter loan amounts, while Union Bank offered the greatest number of previous loan amounts.

**Where are the borrowers located?**

```python
gdf = geopandas.GeoDataFrame(train.loanamount_x,
                             geometry=geopandas.points_from_xy(train.longitude_gps, train.latitude_gps))
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
ax = world.plot(color='white', edgecolor='grey')
gdf.plot(ax=ax, column='loanamount_x', cmap=plt.cm.rainbow_r, legend=True, edgecolor='black')
plt.title('Distribution of Loan Amounts', fontsize=20)
```

**Where are the bad debtors located?**

```python
data = train.loc[train['good_bad_flag'].str.contains('Bad')]
gdf = geopandas.GeoDataFrame(data[['good_bad_flag', 'loanamount_x']],
                             geometry=geopandas.points_from_xy(data['longitude_gps'], data['latitude_gps']))
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
ax = world.plot(color='white', edgecolor='grey')
gdf.plot(ax=ax, color='red', edgecolor='black', legend=True)
plt.title('Location of Bad Debtors', fontsize=20)
```

**What of good debtors?**

```python
data = train.loc[train['good_bad_flag'].str.contains('Good')]
gdf = geopandas.GeoDataFrame(data[['good_bad_flag', 'loanamount_x']],
                             geometry=geopandas.points_from_xy(data['longitude_gps'], data['latitude_gps']))
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
ax = world.plot(color='white', edgecolor='grey')
gdf.plot(ax=ax, color='green', edgecolor='black', legend=True)
plt.title('Location of Good Debtors', fontsize=20)
```
